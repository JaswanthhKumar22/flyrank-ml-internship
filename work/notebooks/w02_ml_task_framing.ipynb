## 1. My lane as an ML task (type)

**Task Type:** Ranking

The goal is to rank content based on its likelihood of performing well for users. Instead of simply classifying content into categories, the model predicts which content should appear higher in search results or recommendations. Ranking helps prioritize the most relevant and engaging content for users.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## 2. Target or proxy

The target is the **engagement score** of a content item. Since a direct quality label may not be available, engagement acts as a proxy target.

The proxy can be calculated from observed user interactions such as clicks, impressions, likes, or other engagement-related signals collected from historical data.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## The primary evaluation metric is **NDCG (Normalized Discounted Cumulative Gain)**.

NDCG measures how well the ranking produced by the model matches the ideal ranking. A higher NDCG value indicates better ranking quality.

A value close to **1.0** indicates excellent performance.

##
