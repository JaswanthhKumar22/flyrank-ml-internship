## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

### Decision

The decision is which content pages should be prioritized for review and potential refresh based on their search performance, engagement, and content freshness.

### Who Acts on It?

The ranked recommendations are intended for content managers, SEO specialists, or digital marketing teams, who can review the suggested pages and decide whether to update, expand, rewrite, monitor, or leave them unchanged.

### Action

For the highest-ranked pages, the reviewer can:
- Refresh outdated content.
- Improve titles or metadata to increase CTR.
- Expand thin content.
- Improve user engagement.
- Monitor pages that require observation rather than immediate changes.

### Cost of a Wrong Recommendation

A false positive means recommending a healthy page for review, which wastes time and resources.

A false negative means failing to identify a page that is genuinely declining, allowing traffic and user engagement to continue decreasing before corrective action is taken.

### Why Machine Learning?

This problem is more than simply training a model. Multiple observable signals—such as impressions, clicks, CTR, average position, content age, and engagement—interact in complex ways. Machine learning can combine these signals to rank pages more effectively than a single rule, while still supporting human decision-making rather than replacing it.
